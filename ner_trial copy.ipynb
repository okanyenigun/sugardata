```python
import pickle
import sys
sys.path.append('/Users/okanyenigun/Desktop/codes/reals/sugardata/')

from dotenv import load_dotenv

load_dotenv()

import sugardata as su

from datasets import load_dataset
```

```python
examples = [
    {
        "text": "Last year, Arsenal won the First Division.", 
        "ner_tags": {"Arsenal": "organization", "First Division": "organization"}
    },
    {
        "text": "Lionel Messi scored a goal.", 
        "ner_tags": {"Lionel Messi": "person"}
    },
    {
        "text": "He lives in Miami.", 
        "ner_tags": {"Miami": "location"}
    }
]
```

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-multilingual-cased")

response = su.localize_ner_data(examples=examples, target_language="tr", verbose=True, tokenizer=tokenizer)
```

```python
for resp in response:
    print(resp)
    print()
```
